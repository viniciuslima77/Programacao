```python
# Escreva um programa que faça2 cálculos aritméticos

# Entrada

a = int (input("digite o primeiro número: "))

b = int ( input("digite o segundo número: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_inteira = a // b # Usar barras duplas "//" para divisões inteiras
resto = a % b # O símbolo de % é igual ao resto da divisão
# Saída


print ("|--------------------CALCULADORA SIMPLES-------------------|")
print ("                                                          |")
print ("|---------------------------SOMA--------------------------|")
print ("                                                          |")
print (f" o resultado de {a} mais {b} é igual a {soma}")
print ("                                                          |")
print ("|------------------------SUBTRAÇÃO------------------------|")
print ("                                                          |")
print (f" o resultado de {a} menos {b} é igual a {subtracao}")
print ("                                                          |")
print ("|-----------------------MULTIPLICAÇÃO---------------------|")
print ("                                                          |")
print (f" o resultado de {a} vezes {b} é igual a {multiplicacao}")
print ("                                                          |")
print ("|--------------------------DIVISÃO------------------------|")
print ("                                                          |")
print (f" o resultado de {a} dividido por {b} é igual a {divisao}")
print ("                                                          |")
print ("|-----------------------DIVISÃO INTEIRA-------------------|")
print ("                                                          |")
print (f" o resultado de {a} dividido por {b} é igual a {div_inteira}")
print ("                                                          |")
print ("|---------------------------RESTO-------------------------|")
print ("                                                          |")
print (f" o resto de {a} dividido por {b} é igual a {resto}")
print ("|---------------------------------------------------------|")
```
