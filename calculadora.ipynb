**ATIVIDADE**

Escreva um programa que receba a temperatura em graus celsius e converte para fahrenheit
# Entrada

Graus celsius

# Processamento

F° = C° x (9/5) + 32

# Saída

Mostrar graus em fahrenheit

```python
# Entrada

Celsius = int (input(f"digite a temperatura em graus celsius:  "))

# Processamento

Fahrenheit = Celsius * (9/5) + 32

# Saída

(f"a temperatura em fahrenheit é: {Fahrenheit} F°")
```
