1. Par ou ímpar
Peça um número ao usuário e informe se ele é par ou ímpar.

```python
numero01 = int (input ("Digite o número: "))

if numero01 % 2 == 0:
    print ("esse número é par")
else:
    print ("esse número é impar")
```

2. Maior de dois números
Solicite dois números e mostre qual é o maior (ou se são iguais).

```python
numero01 = int (input ("digite um numero: "))
numero02 = int (input ("digite outro numero: "))

if numero01 > numero02:
    print (f"{numero01} é maior que {numero02}")

elif numero02 > numero01:
    print (f"{numero02} é maior que {numero01}")
else:
    print ("esses números são iguais")
```

3. Aprovação de aluno
Peça a nota de um aluno e informe:

o Aprovado (nota ≥ 7)

o Recuperação (5 ≤ nota &lt; 7)

o Reprovado (nota &lt; 5)

```python
nota = float (input ("digite a sua nota:"))

if nota >= 7:
  print("Parabéns, você foi aprovado!")
elif nota >= 5 and nota < 7:
  print("Você ficou de recuperação.")
else:
  print("Você foi reprovado.")
```

4. Classificação de idade
Receba a idade e classifique:

o Criança (0–12)

o Adolescente (13–17)

o Adulto (18–59)

o Idoso (60+)

```python
idade = int (input ("digite sua idade: "))

if 12 > idade > 0:
  print ("Você é criança")
elif 17 > idade > 13:
  print ("Você é adolescente")
elif 59 > idade > 18:
  print ("Você é adulto")
else:
  print ("Você é idoso")
```

7. Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o
resultado.

```python
numero01 = int (input ("digite um numero: "))
numero02 = int (input ("digite outro numero: "))
print ("operações disponíveis:")
print ("adição = +")
print ("subtração = -")
print ("multiplicação = *")
print ("divisão = /")

operacao = input ("digite a operação: ")


if operacao == "+":
  print (numero01 + numero02)
elif operacao == "-":
  print (numero01 - numero02)
elif operacao == "*":
  print (numero01 * numero02)
elif operacao == "/":
  print (numero01 / numero02)
```

Dia da semana
Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

```python
print ("1 = domingo")
print ("2 = segunda-feira")
print ("3 = terça-feira")
print ("4 = quarta-feira")
print ("5 = quinta-feira")
print ("6 = sexta-feira")
print ("7 = sábado")

dia = int (input ("digite o dia da semana: "))

if dia == 1:
  print ("domingo")
elif dia == 2:
  print ("segunda-feira")
elif dia == 3:
  print ("terça-feira")
elif dia == 4:
  print ("quarta-feira")
elif dia == 5:
  print ("quinta-feira")
elif dia == 6:
  print ("sexta-feira")
elif dia == 7:
  print ("sábado")
else:
  print ("insira um dia válido")
```

Ano bissexto
Solicite um ano e informe se ele é bissexto ou não.

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

10. Desconto em compras
Peça o valor de uma compra:

 ≥ 100: 10% de desconto

 ≥ 50 e &lt; 100: 5% de desconto

 &lt; 50: sem desconto

```python
compra = float (input ("digite o valor da compra: "))
if compra >= 100:
  print ("você ganhou 10% de desconto")
  print (f" o total agora é: R$ {compra - (compra * 0.10)}")
elif 100 > compra >= 50:
  print ("você ganhou 5% de desconto")
  print (f" o total agora é: R$ {compra - (compra * 0.05)}")
else:
  print ("você não ganhou desconto")
```
