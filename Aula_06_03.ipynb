*Calcular o dobro dos livros*

# **Processamento**

*calcular a quantidade de prateleiras*

*calcular o total e o dobro dos livros*

# **Saída**

```python
# Entrada

prateleiras = 5
livros = 15

# Processamento

total_de_livros = prateleiras * livros
dobro = total_de_livros * 2

# Saída

print (f"o total do dobro de livros é: {dobro}")
```
